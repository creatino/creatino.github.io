```python
from polygeohasher import polygeohasher
import geopandas as gpd
gdf = gpd.read_file('../content/gdrive/My Drive/Colab_Notebooks/NTIA/GEOHASH/US_States.json')
```

```python
# Assuming 'geohash_level' is defined somewhere above
geohash_level = 5  # Replace with your desired geohash level

# Instantiate the polygeohasher object
polygeohasher_obj = polygeohasher.Polygeohasher(gdf)

# Call the function using the object, passing geohash_level as an integer
primary_df = polygeohasher_obj.create_geohash_list(geohash_level=geohash_level)  # Pass geohash_level as a keyword argument
print(primary_df)
```

```python
# Print number of geohashes in each state

# Calculate the number of geohashes for each state
geohash_counts = primary_df['geohash_list'].apply(len)

# Print the number of geohashes for each state
for index, row in primary_df.iterrows():
    state_name = row['NAME']
    count = geohash_counts[index]
    print(f"Number of geohashes in {state_name}: {count}")
```

```python
import pandas as pd
import geohash # Import the geohash module

# Load the USC Sites data
sites_df = pd.read_excel('../content/gdrive/My Drive/Colab_Notebooks/NTIA/GEOHASH/USC_Sites.xlsx')

# Extract SiteNum, Latitude, and Longitude
site_data = sites_df[['SiteNum', 'Latitude', 'Longitude']].values.tolist()

# Map coordinates to geohashes
geohash_mapping = {}
for site_num, lat, lon in site_data:
    geohash_str = geohash.encode(lat, lon, precision=geohash_level)
    geohash_mapping[site_num] = geohash_str

# Print the geohash mapping
for site_num, geohash_str in geohash_mapping.items():
    print(f"SiteNum: {site_num}, Geohash: {geohash_str}")
```

```python
# Print total number of geohashes and SiteNum's

# Total number of geohashes
total_geohashes = sum(int(count) for count in geohash_counts.values()) # Convert counts to integers before summing
print(f"Total number of geohashes: {total_geohashes}")

# Total number of SiteNums
total_sitenums = len(geohash_mapping)
print(f"Total number of SiteNums: {total_sitenums}")
```

```python
# Print the number of SiteNum's contained in each Geohash

# Count SiteNums per geohash
geohash_counts = {}
for site_num, geohash_str in geohash_mapping.items():
    if geohash_str not in geohash_counts:
        geohash_counts[geohash_str] = 0
    geohash_counts[geohash_str] += 1

# Print the counts
for geohash_str, count in geohash_counts.items():
    print(f"Geohash: {geohash_str}, Number of SiteNums: {count}")
```

```python
# Print the number of Geohash's containing multiple SiteNums

# Count geohashes with multiple SiteNums
multiple_sitenums_count = sum(1 for count in geohash_counts.values() if count > 1)
print(f"Number of Geohash's containing multiple SiteNums: {multiple_sitenums_count}")
```

```python
# List the Geohash's containing multiple SiteNums and the SiteNums

# Count geohashes with multiple SiteNums and list them
multiple_sitenums_geohashes = []
for geohash_str, count in geohash_counts.items():
    if count > 1:
        # Find the SiteNums associated with this geohash
        sitenums_in_geohash = [site_num for site_num, gh in geohash_mapping.items() if gh == geohash_str]
        multiple_sitenums_geohashes.append((geohash_str, sitenums_in_geohash))

# Print the geohashes with multiple SiteNums and the corresponding SiteNums
print("Geohashes with multiple SiteNums:")
for geohash_str, sitenums in multiple_sitenums_geohashes:
    print(f"Geohash: {geohash_str}, SiteNums: {', '.join(map(str, sitenums))}")
```

```python
# For every Geohash containing multiple SiteNums, select and keep only 1 SiteNums and delete the  rest of the SiteNums. List the modified Geohashs, corresponding SiteNums, number of SiteNums, and corresponding Latiitude and Longitude

modified_geohash_mapping = geohash_mapping.copy()
for geohash_str, sitenums in multiple_sitenums_geohashes:
    # Keep only the first SiteNum
    sitenum_to_keep = sitenums[0]
    for site_num in sitenums[1:]:
        del modified_geohash_mapping[site_num]

# Print the modified geohash mapping
print("Modified Geohash Mapping:")
for site_num, geohash_str in modified_geohash_mapping.items():
    print(f"SiteNum: {site_num}, Geohash: {geohash_str}")

# Count SiteNums per geohash in the modified mapping
modified_geohash_counts = {}
for site_num, geohash_str in modified_geohash_mapping.items():
    if geohash_str not in modified_geohash_counts:
        modified_geohash_counts[geohash_str] = 0
    modified_geohash_counts[geohash_str] += 1

# Print the counts for the modified mapping
print("\nNumber of SiteNums per Geohash (Modified):")
for geohash_str, count in modified_geohash_counts.items():
    print(f"Geohash: {geohash_str}, Number of SiteNums: {count}")

# Print the modified Geohashs, corresponding SiteNums, number of SiteNums, and corresponding Latitude and Longitude
print("\nModified Geohashes with SiteNums, Counts, Latitude, and Longitude:")
for geohash_str, count in modified_geohash_counts.items():
    # Find the SiteNum associated with this geohash
    site_num = next((site_num for site_num, gh in modified_geohash_mapping.items() if gh == geohash_str), None)
    if site_num is not None:
        # Find the Latitude and Longitude for this SiteNum
        latitude = sites_df.loc[sites_df['SiteNum'] == site_num, 'Latitude'].iloc[0]
        longitude = sites_df.loc[sites_df['SiteNum'] == site_num, 'Longitude'].iloc[0]
        print(f"Geohash: {geohash_str}, SiteNum: {site_num}, Count: {count}, Latitude: {latitude}, Longitude: {longitude}")
```

```python
# Print total number of geohashes and SiteNum's

# Total number of geohashes
total_geohashes = sum(int(count) for count in geohash_counts.values()) # Convert counts to integers before summing
print(f"Total number of geohashes: {total_geohashes}")

# Total number of SiteNums
total_sitenums = len(geohash_mapping)
print(f"Total number of SiteNums: {total_sitenums}")
```

```python
# For every Geohash containing more than one SiteNum, pick a random SiteNum to keep and eliminate the rest. list the Geohashes and corresponding SiteNum, Latitude and Longitude

import random

# Filter geohashes with more than one SiteNum
geohashes_to_process = {k: v for k, v in geohash_counts.items() if v > 1}

# Process geohashes and select a random SiteNum
final_mapping = {}
for geohash_str, count in geohashes_to_process.items():
  # Get all SiteNums for this geohash
  site_nums_in_geohash = [site_num for site_num, gh in geohash_mapping.items() if gh == geohash_str]
  # Choose a random SiteNum
  chosen_site_num = random.choice(site_nums_in_geohash)
  final_mapping[geohash_str] = chosen_site_num

# Get Latitude and Longitude for chosen SiteNums
final_data = []
for geohash_str, site_num in final_mapping.items():
  row = sites_df[sites_df['SiteNum'] == site_num].iloc[0]
  final_data.append({
      'Geohash': geohash_str,
      'SiteNum': site_num,
      'Latitude': row['Latitude'],
      'Longitude': row['Longitude']
  })

# Print the final data
final_df = pd.DataFrame(final_data)
print(final_df)
```
